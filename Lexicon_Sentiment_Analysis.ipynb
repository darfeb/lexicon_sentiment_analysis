```python
#library for punctuation
import string
string.punctuation

#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
df['content_clean']= df['content'].apply(lambda x:remove_punctuation(x))
df.head()
```

```python
#lowering text
df['lower_text']= df['content_clean'].apply(lambda x: x.lower())
df.head()
```

```python
import re
#remove number
df['clear_number'] = df['lower_text'].apply(lambda x: re.sub(r'\d+', '', x))
df.head()
```

```python
#remove hastag
df['clear_hastag'] = df['clear_number'].apply(lambda x: re.sub(r'#\w+', '', x))
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#remove stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

df['no_stopwords'] = df['clear_hastag'].apply(remove_stopwords)
df.head()
```

```python
import re
#remove emojis
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emojis'] = df['no_stopwords'].apply(remove_emojis)
df.head()
```

```python
#remove whitespace
def remove_whitespace(text):
    return ' '.join(text.split())

df['no_whitespace'] = df['no_emojis'].apply(remove_whitespace)
df.head()
```

```python
# check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#tokenized
def tokenization(text):
    tokens = re.split(r'\W+', text)
    return tokens
#applying function to the column
df['tokenized']= df['no_whitespace'].apply(lambda x: tokenization(x))
df.head()
```

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer

#defining the object for stemming
porter_stemmer = PorterStemmer()

#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text

df['stemming_text'] = df['tokenized'].apply(lambda x: stemming(x))
df.head()
```

```python
import spacy
import pandas as pd

#initialize SpaCy dan load english model
nlp = spacy.load('en_core_web_sm')

#function for lemmatization using SpaCy
def lemmatization(tokens):
    doc = nlp(' '.join(tokens))
    lemmas = [token.lemma_ for token in doc]
    return lemmas

#function lemmatization from 'tokenized' to new column 'lemmatization_text'
df['lemmatization_text'] = df['tokenized'].apply(lambda x: lemmatization(x))
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
```

```python
#initialize lists to store sentiment scores
sentiments = []

#initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

#iterate over the texts and apply sentiment analyzer
for lemmatization_text in tqdm(df['lemmatization_text']):
    # Join the list of lemmatized tokens back into a single string
    lemmatization_text_str = ' '.join(lemmatization_text)
    scores = sia.polarity_scores(lemmatization_text_str)
    sentiments.append(scores)
    print("Sentiments Scores:", scores)

#create a DataFrame from the sentiment scores
sentiment_df = pd.DataFrame(sentiments)

#save the DataFrame
output_file = 'sentiment_scores.xlsx'
sentiment_df.to_excel(output_file, index=False)
```

```python
# Add the sentiment scores to the original DataFrame
df = pd.concat([df, sentiment_df], axis=1)

# Check the columns of the combined DataFrame
print(df.columns)
df.head()
```

```python
#add sentiment scores to the DataFrame
df['sentiment_scores'] = sentiments

#function to assign sentiment label based on neg, neu, and pos scores
def get_sentiment_label(scores):
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

#create the 'label_text' column
df['label_text'] = df['sentiment_scores'].apply(get_sentiment_label)
df.head()
```
